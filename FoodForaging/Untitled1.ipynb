```python
rows = []
for i in range(16):
    base_two_str = str(bin(i))[2:]
    zeros_to_append = 4 - len(base_two_str)
    for _ in range(zeros_to_append):
        base_two_str = '0' + base_two_str
    row_part = f'State {i}: |'
    #base_two_str = "".join(reversed(base_two_str))
    for binary in base_two_str:
        if binary == '0':
            row_part = row_part + ' ______ | '
        else: 
            row_part = row_part + ' GREEN | '
    rows.append(row_part)
cols = ["STRAIGHT", "LEFT", "RIGHT"]
df = pd.DataFrame(data=food_q_matrix, index=rows, columns=cols)
df.style.highlight_max(color = 'lightblue', axis = 1)
```

```python
rows = []
food_q_matrix[0][2] = 1
food_q_matrix[1][2] = 17
#food_q_matrix[2][0] = 20

for i in range(16):
    base_two_str = str(bin(i))[2:]
    zeros_to_append = 4 - len(base_two_str)
    for _ in range(zeros_to_append):
        base_two_str = '0' + base_two_str
    row_part = f'State {i}: |'
    if base_two_str[0] == '0':
        row_part = row_part + ' NOTACT | '
    else: 
        row_part = row_part + ' ACTIVE | '
    base_two_str = base_two_str[1:]
    #base_two_str = "".join(reversed(base_two_str))
    for binary in base_two_str:
        if binary == '0':
            row_part = row_part + ' ______ | '
        else: 
            row_part = row_part + ' GREEN | '
    rows.append(row_part)
cols = ["STRAIGHT", "LEFT", "RIGHT"]
df = pd.DataFrame(data=food_q_matrix, index=rows, columns=cols)
df.style.highlight_max(color = 'lightblue', axis = 1)
```
