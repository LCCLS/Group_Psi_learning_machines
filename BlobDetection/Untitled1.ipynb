```python
import numpy as np
import pandas as pd
```

```python
q_matrix = np.load('matrices_kindofok/q_matrix_new_final.npy')

q_matrix[0][2] = 1
q_matrix[15][0] = 48
q_matrix[30][0] = 50
q_matrix[5][0] = 42 # Maybe remove?
q_matrix[31][0] = 57

q_matrix[11][1] = 41
q_matrix[11][2] = 51

q_matrix[4][0] = 64
q_matrix[4][1] = 39

q_matrix[23][0] = 43
q_matrix[24][1] = 37
```

```python
rows = []
for i in range(32):
    base_two_str = str(bin(i))[2:]
    zeros_to_append = 5 - len(base_two_str)
    for _ in range(zeros_to_append):
        base_two_str = '0' + base_two_str
    row_part = f'State {i}: |'
    #base_two_str = "".join(reversed(base_two_str))
    for binary in base_two_str:
        if binary == '0':
            row_part = row_part + ' ______ | '
        else: 
            row_part = row_part + ' GREEN | '
    rows.append(row_part)
cols = ["STRAIGHT", "LEFT", "RIGHT"]
df = pd.DataFrame(data=q_matrix, index=rows, columns=cols)
df.style.highlight_max(color = 'lightblue', axis = 1)
```
